# 次元削減

データの次元を減らすこと。

```python
import numpy as np
import matplotlib.pyplot as plt
```

次元を減らすという概念を説明する。

ある$l$次元のデータ空間$\R^l$に分布するデータセット$\mathcal X$について考える。データ$\boldsymbol x_n\in\mathcal X$は$l$個の値$x_1, x_2, \cdots, x_l$によって一意に表せる。しかし、扱うデータの分布によっては$l$より少ない個数で$\boldsymbol x_n$を一意に表せる。

例を示そう。2次元空間$\R^2$にデータ$\boldsymbol x_n=(x_1, x_2)$が存在し、それが$3x_1-4x_2=0$を満たす超平面（直線）**のみ**に分布する場合を考える。

```python
n = 15
a = 3/4
x1 = np.random.random(n)
x2 = a * x1
plt.scatter(x1, x2, color='magenta');
```

このとき、$x_1$を定めれば$x_2$が一意に定まる（逆も然り）ため、$\R^2$内のデータ$\boldsymbol x_n$を表すのに必要な値は一つで良いことになる。これは、2次元空間内のデータが軸を減らした1次元空間で完全に表せることを示している。

では、もう少し現実的な例について考える。ある2次元のデータが$3x_1-4x_2=0$を満たす超平面の**近く**に分布する場合を考える。

```python
x2 += 0.05 * np.random.randn(n)
plt.scatter(x1, x2, color='magenta');
```

この場合、先ほどのように1次元空間で完全に表すことはできない。しかし、**だいたい**を表すことはできる。  
このデータに最も適合する直線を求め、その直線上のどの辺りに位置するかを記述すれば、2次元空間内のデータを1次元空間内の位置で近似的に表現できる。

このように、本来の次元よりも少ない次元でデータセットを近似的に表現することを次元削減という。
